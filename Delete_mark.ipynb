```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import pandas as pd 
import seaborn as sns 
import sklearn 
```

```python
url = "https://raw.githubusercontent.com/samarkinav/ML/main/students.csv"
data_raw = pd.read_csv(url)
```

```python
data_raw.shape
```

```python
data_raw.drop([395], inplace = True)
```

```python
data_raw.shape
```

```python
data_raw['G3'] = data_raw['G3'].astype(str).astype(int)
```

```python
def func(G3):
    if G3 < 10:
        return 'fail'
    else:
        return 'passed'
    
data_raw['Mark'] = data_raw.apply(lambda x: func(x.G3), axis = 1)
```

```python
data_raw.head()
```

```python
data_raw.drop('G1', axis = 1, inplace = True)
```

```python
data_raw.drop('G2', axis = 1, inplace = True)
```

```python
data_raw.drop('G3', axis = 1, inplace = True)
```
